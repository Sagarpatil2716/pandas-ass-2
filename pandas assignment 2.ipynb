```python
import pandas as pd

# URL of the dataset
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'

# Read the dataset using '|' as the separator
users = pd.read_csv(url, sep='|')

# First 10 entries
print("--------First 10 entries-------")
print(users.head(10))

# Last 10 entries
print("--------Last 10 entries-------")
print(users.tail(10))

# Number of observations
num_observations = users.shape[0]
print("Number of observations in the dataset:", num_observations)

# Number of columns
num_columns = users.shape[1]
print("Number of columns in the dataset:", num_columns)

# Name of all columns
print("Name of all columns:")
print(users.columns)

# Dataset index
print("Dataset indexed by:", users.index.name)

# Data type of each column
print("Data type of each column:")
print(users.dtypes)

# Occupation column
print("Occupation column:")
print(users['occupation'])

# Number of different occupations
num_occupations = users['occupation'].nunique()
print("Number of different occupations:", num_occupations)

# Most frequent occupation
most_frequent_occupation = users['occupation'].value_counts().idxmax()
print("Most frequent occupation:", most_frequent_occupation)

# Describe all columns
print("Description of all columns:")
print(users.describe())

# Describe only the occupation column
print("Description of the occupation column:")
print(users['occupation'].describe())

# Mean age of users
mean_age = users['age'].mean()
print("Mean age of users:", mean_age)

# Age with the least occurrence
least_occurrence_age = users['age'].value_counts().idxmin()
print("Age with the least occurrence:", least_occurrence_age)
```
